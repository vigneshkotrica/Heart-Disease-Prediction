```python
import pandas as pd
from sklearn.cluster import KMeans

df = pd.read_csv("students.csv")

X = df[["CGPA","Attendence"]]

model = KMeans(
    n_clusters=3,
    random_state=42
)

model.fit(X)

df["Cluster"] = model.labels_

print(df)
```

```python
import matplotlib.pyplot as plt

plt.scatter(
    df["CGPA"],
    df["Attendence"],
    c=df["Cluster"]
)

plt.xlabel("CGPA")
plt.ylabel("Attendance")
plt.title("Student Clusters")

plt.show()
```
